# **Problem 1**

mtcars.csv – Dataset extracted from the 1974 Motor Trend US magazine comprising fuel consumption and 10 aspects (attributes) of automobile design and performance for 32 automobiles is a famous problem dataset used in machine learning and data analysis. We will use this dataset to perform data abstraction, slicing, dicing and basic analysis using Pandas
Series and DataFrame, as taught in the class.

One of the main purpose of this data was to assist decision making on which car to purchase during the 1973 Oil Crisis which began in October 1973, when the members of the Organization of Arab Petroleum Exporting Countries proclaimed an oil embargo and the affected countries were Canada Japan, Netherlands, United Kingdom and United States and later extended
to Portugal, Rhodesia and South Africa. By the end of the embargo in March 1974, the oil price had risen nearly 400%, from \$3 per barrel to nearly \$12 globally; US prices were significantly higher. The embargo caused an oil
crisis, or &quot;shock&quot;, with many short and long-term effects on global politics and the global economy. It was famously called the “First Oil Shock”.

### 1. We will perform data analysis using Pandas to determine the best car to buy based on the tasks given below.

1.a Reading the csv file using Pandas. 

```
# Type solution here
frame = pd.read_csv("mtcars.csv")
print(frame)
```

1.b Making the column Car Name the index of frame

```
frame.set_index("Car Name", inplace=True)
print(frame)
```

1.c Creating a DataFrame using the attributes: 'cyl', 'gear', 'hp', 'mpg'

```
df = frame[['cyl', 'gear', 'hp', 'mpg']]
df.columns=['Cylinders', 'Gear', 'Horsepower', 'Miles Per Gallon']
print(df)
```

1.d Identifying cars with ‘Horsepower’ greater than or equal to 110. Adding a separate column called ‘Powerful’ to the DataFrame which holds the result of cars with ‘Horsepower’ greater than or equal to 110

```
df['Powerful'] = (df['Horsepower'] >= 110)
print(df)
```

1.e Storing the dataframe in another variable. Printing the new DataFrame with the column ‘Horse Power’ deleted

```
df_new = df.copy()
del df_new['Horsepower']
print(df_new)
print(df)
```

1.f Listing cars with 'Miles Per Gallon' greater than 25.0 and sorting the cars in descending order of 'Horsepower'

```
df_powerful = df[df['Miles Per Gallon'] > 25.0].sort_values('Horsepower', ascending=False)
print(df_powerful)
```

1.g Filtering out the car that is powerful and has the highest Miles Per Gallon.

```
print(df_powerful[df_powerful['Powerful'] == True].sort_values('Miles Per Gallon', ascending=False).iloc[0:1])
```

### 1. Preparation

1.a Reading the dataset into a pandas dataframe. columns: Date, AveragePrice, Total Volume. 

```
avocado = pd.read_csv('avocado.csv')[['Date', 'AveragePrice', 'Total Volume']]
print(avocado)
```

1.b Converting Date column to a timestamp using datetime.

```
from datetime import datetime
avocado['Date'] = pd.to_datetime(avocado['Date'])
print(avocado)
```

2. Plotting

2.a Creating a figure with 4 subplots.

```
import matplotlib.pyplot as plt

fig = plt.figure()
p1 = fig.add_subplot(2, 2, 1)
p2 = fig.add_subplot(2, 2, 2)
p3 = fig.add_subplot(2, 2, 3)
p4 = fig.add_subplot(2, 2, 4)
fig.set_figheight(15)
fig.set_figwidth(15)
```

2.b Sorting avocado by Date inplace in ascending order.

```
avocado = avocado.sort_values(by='Date', ascending=True)
print(avocado)
```

2.c Plotting the average price of avocados over time in subplot 1.

```
import matplotlib.pyplot as plt

fig = plt.figure()
p1 = fig.add_subplot(2, 2, 1)
p2 = fig.add_subplot(2, 2, 2)
p3 = fig.add_subplot(2, 2, 3)
p4 = fig.add_subplot(2, 2, 4)
fig.set_figheight(15)
fig.set_figwidth(15)

p1.scatter(avocado['Date'], avocado['AveragePrice'])
p1.set_ylabel('Average Price')
labels = p1.set_xticklabels(['2015-01','2015-05','2015-09','2016-01','2016-05','2016-09','2017-01','2017-05','2017-09','2018-01','2018-05'], rotation=90, fontsize='small') # xtick labels
plt.show()
```

2.d Plotting the total volume of avocados sold over time in subplot 2.

```
fig = plt.figure()
p1 = fig.add_subplot(2, 2, 1)
p2 = fig.add_subplot(2, 2, 2)
p3 = fig.add_subplot(2, 2, 3)
p4 = fig.add_subplot(2, 2, 4)
fig.set_figheight(15)
fig.set_figwidth(15)

p1.scatter(avocado['Date'], avocado['AveragePrice'])
p1.set_ylabel('Average Price')
labels = p1.set_xticklabels(['2015-01','2015-05','2015-09','2016-01','2016-05','2016-09','2017-01','2017-05','2017-09','2018-01','2018-05'], rotation=90, fontsize='small') # xtick labels

p2.scatter(avocado['Date'], avocado['Total Volume'])
p2.set_ylabel('Total Volume')
labels = p2.set_xticklabels(['2015-01','2015-05','2015-09','2016-01','2016-05','2016-09','2017-01','2017-05','2017-09','2018-01','2018-05'], rotation=90, fontsize='small') # xtick labels

plt.show()
```

2.e The plots are cluttered because there are many dates in the dataframe and there are several transactions on the same date.

To address this, I aggregated the volume and price by date.

```
avocado['TotalRevenue'] = avocado['AveragePrice']*avocado['Total Volume']
avocado1 = avocado.groupby('Date').sum()
avocado1['AveragePrice'] = avocado1['TotalRevenue']/avocado1['Total Volume']
print(avocado1)
```

2.f Plotting the average price of avocado1 over time in subplot 3.

```
fig = plt.figure()
p1 = fig.add_subplot(2, 2, 1)
p2 = fig.add_subplot(2, 2, 2)
p3 = fig.add_subplot(2, 2, 3)
p4 = fig.add_subplot(2, 2, 4)
fig.set_figheight(15)
fig.set_figwidth(15)

p1.scatter(avocado['Date'], avocado['AveragePrice'])
p1.set_ylabel('Average Price')
labels = p1.set_xticklabels(['2015-01','2015-05','2015-09','2016-01','2016-05','2016-09','2017-01','2017-05','2017-09','2018-01','2018-05'], rotation=90, fontsize='small') # xtick labels

p2.scatter(avocado['Date'], avocado['Total Volume'])
p2.set_ylabel('Total Volume')
labels = p2.set_xticklabels(['2015-01','2015-05','2015-09','2016-01','2016-05','2016-09','2017-01','2017-05','2017-09','2018-01','2018-05'], rotation=90, fontsize='small') # xtick labels

p3.plot(avocado1.index, avocado1['AveragePrice'], marker='o')
p3.set_ylabel('Average Price')
labels = p3.set_xticklabels(['2015-01','2015-05','2015-09','2016-01','2016-05','2016-09','2017-01','2017-05','2017-09','2018-01','2018-05'], rotation=90, fontsize='small') # xtick labels

plt.show()
```
